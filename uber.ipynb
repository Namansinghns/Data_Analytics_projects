```
import pandas as pd
import matplotlib.pyplot as plt
```

```
ride = pd.read_csv("/content/ncr_ride_bookings.csv")
ride
```

```
ride[ride.duplicated()]
```

```
ride.drop_duplicates()
```

```
ride.info()
```

```
ride.describe()
```

```
ride.isnull().sum()
```

```
total_revenue = ride["Booking Value"].sum()    # total revenue
total_revenue
```

```
ride["Date"] = pd.to_datetime(ride["Date"])   # total revenue by weeks
ride["week"] = ride["Date"].dt.day_name()
weekly_revenue = ride.groupby("week")["Booking Value"].sum().reindex(["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"])
weekly_revenue
```

```
plt.plot(weekly_revenue/1e6)
plt.xlabel("Week")
plt.ylabel("Total Revenue")
plt.title("Total Revenue by Week")
plt.show()
```

```
ride["date"] = pd.to_datetime(ride["Date"])       # total revenue by months
ride["month"] = ride["date"].dt.month_name()
monthly_revenue = ride.groupby("month")["Booking Value"].sum().reindex(["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"])
monthly_revenue
```

```
plt.plot(monthly_revenue/1e6 , marker = "^")
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.title("Total Revenue by Month")
plt.xticks(rotation = 45)
plt.show()
```

```
ride_values = ride.groupby("Vehicle Type")["Booking Value"].sum().sort_values(ascending=False)    # total revenue by vehicle
print(ride_values)
```

```
plt.bar(ride_values.index, ride_values.values/1e7)
plt.xlabel("Vehicle Type")
plt.ylabel("Total Revenue crore")
plt.title("Total Revenue by Vehicle Type")
plt.xticks(rotation = 45)
plt.show()
```

```
total_payment = ride.groupby("Payment Method")["Booking Value"].sum().sort_values(ascending=False)   # total revenue by payment method
total_payment
```

```
plt.bar(total_payment.index, total_payment.values/1e7)
plt.xlabel("Payment Method")
plt.ylabel("Total Revenue crore")
plt.title("Total Revenue by Payment Method")
plt.show()
```

```
booking_status = ride["Booking Status"].value_counts()     # values count by status
booking_status
```

```
plt.pie(booking_status.values, labels=booking_status.index, autopct="%1.1f%%")
plt.title("Booking Status")
plt.show()
```

```
rider_distance = ride.groupby("Vehicle Type")["Ride Distance"].sum()   # total ride distance by vehicle
rider_distance
```

```
average = ride.groupby("Vehicle Type")["Ride Distance"].mean()   #  average ride distance
average
```

```
customer_rating = ride.groupby("Vehicle Type")["Customer Rating"].mean()
customer_rating
```

```
plt.pie(customer_rating.values, labels=customer_rating.index, autopct="%1.1f%%")   # average rating by customers
plt.title("Customer Rating")
plt.show()
```

```
driver_rating = ride.groupby("Vehicle Type")["Driver Ratings"].mean()
driver_rating
```

```
plt.pie(driver_rating.values, labels=driver_rating.index, autopct="%1.1f%%")   # average rating by drivers
plt.title("Driver Rating")
plt.show()
```
