```
import pandas as pd
import matplotlib.pyplot as plt
```

```
store = pd.read_csv("/content/Sample - Superstore.csv",encoding="latin1")
store
```

```
total_profit = store["Profit"].sum()    #  total profit
print(round(total_profit,2))
```

```
total_discount = store["Discount"].sum()   # total discount
print(round(total_discount,2))
```

```
total_revenue = store["Sales"].sum()   # total revenue
print(round(total_revenue,2))
```

```
store["Order Date"] = pd.to_datetime(store["Order Date"])    # revenue by week
store["week"] = store["Order Date"].dt.day_name()
weekly_profit = store.groupby("week")["Sales"].sum().reindex(["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"])
weekly_profit
```

```
colors = ["blue"]*len(weekly_profit)                    #  this will show lowest and highest by week
colors[4] = "green"
colors[0] = "green"
colors[2] = "red"
plt.bar(weekly_profit.index, weekly_profit.values,color = colors)
plt.xlabel("Week")
plt.ylabel("Profit")
plt.title("Weekly Profit")
plt.xticks(rotation = 45)
plt.show()
```

```
store["Order Date"] = pd.to_datetime(store["Order Date"])    #  monthly revenue
store["Month"] = store["Order Date"].dt.month_name()
monthly_profit = store.groupby("Month")["Sales"].sum().reindex(["January","February","March","April","May","June","July","August","September","October","November","December"])
monthly_profit
```

```
colors = ["Blue"]*len(monthly_profit)
colors[1]  = "red"
colors[-2] = "green"
plt.bar(monthly_profit.index, monthly_profit.values, color = colors)
plt.xlabel("Month")
plt.ylabel("Profit")
plt.title("Monthly Profit")
plt.xticks(rotation = 45)
plt.show()
```

```
store["Order Date"] = pd.to_datetime(store["Order Date"])   # yearly revenue
store["Year"] = store["Order Date"].dt.year
yearly_profit = store.groupby("Year")["Sales"].sum()
yearly_profit
```

```
colors = ["blue"]*len("yearly_profit")
colors[1] = "red"
colors[3] = "green"


year = store["Year"].astype(int).unique()        #  this force years on x-axis
plt.bar(yearly_profit.index, yearly_profit.values, color = colors)
plt.xlabel("Year")
plt.ylabel("Profit")
plt.title("Yearly Profit")
plt.xticks(year)

plt.show()
```

```
ship_mode = store.groupby("Ship Mode")["Category"].value_counts()   # ship by category
ship_mode
```

```
colors = ["blue","green","yellow","red","grey","pink"]
ship_mode.plot(kind = "barh" , color = colors)
plt.xlabel("conuts")
plt.ylabel("category")
plt.title("category by ship")
plt.show()
```

```
category = store.groupby("Category")["Sub-Category"].value_counts()  # sub category count
category
```

```
colors = ["blue","green","yellow","red","grey"]
category.plot(kind ="barh" , color = colors )
plt.xlabel("Sub-Category")
plt.ylabel("sub-category-count")
plt.title("Sub-Category by Category")
plt.show()
```

```
profit_category = store.groupby(["Category","Sub-Category"])["Sales"].sum()   # profit by category wise
profit_category
```

```
colors = ["green"  if val >= 300000 else "red" if val <= 100000   else "blue" for val in profit_category]

profit_category.plot(kind = "barh" , color = colors)
plt.xlabel("Profit")
plt.ylabel("Sub-Category")
plt.title("Profit by Sub-Category")
plt.show()
```

```
quantity = store.groupby(["Category","Sub-Category"])["Quantity"].sum()   # quantity by category
quantity
```

```
colors = ["green" if val >= 3500 else "blue" for val in quantity]

quantity.plot(kind = "barh" , color = colors)
plt.xlabel("Quantity")
plt.ylabel("Sub-Category")
plt.title("Quantity by Sub-Category")
plt.show()
```

```
segement = store.groupby("Segment")["Sales"].sum()   # sales by segment
pd.set_option("display.float_format","{:.2f}".format)
segement
```

```
colours = ["red","green","orange"]
startangle = 90    # rotation
segement.plot(kind = "pie" , autopct = "%1.1f%%" , color = colors)
plt.title("Sales by Segment")
plt.show()
```

```
country = store.groupby("Country")["Sales"].sum()   # sales by country
country
```

```
city = store.groupby("City")["Sales"].sum().sort_values(ascending=False)   # sales by city
city.head(15)
```

```
topcity = city.sort_values(ascending = False).head(10)
colors = ["green" if val == topcity.max()else "red" if val <= 50000 else "blue" for val in topcity]
topcity.plot(kind = "barh", color = colors)
plt.xlabel("Sales")
plt.ylabel("City")
plt.title("Sales by City")
plt.show()
```

```
state = store.groupby(["Region","State"])["Sales"].sum()    #  sales by region and state
state
```

```
shipmode = store.groupby("Ship Mode")["Sales"].sum()   # sales by ship mode
shipmode
```

```
store["Ship Date"] = pd.to_datetime(store["Ship Date"])
store["Month"] = store["Ship Date"].dt.month_name()
store["Day"] = store["Ship Date"].dt.day_name()
store["Year"] = store["Ship Date"].dt.year

```

```
store["Order Date"] = pd.to_datetime(store["Order Date"])
store["Month"] = store["Order Date"].dt.month_name()
store["Day"] = store["Order Date"].dt.day_name()
store["Year"] = store["Order Date"].dt.year
```

```
store["shipping_days"] = (store["Ship Date"] - store["Order Date"]).dt.days   # average shipping time taken
ship_day = store.groupby(["Category","Sub-Category"])["shipping_days"].mean()
ship_day
```

```
shipping_days = store.groupby(store["Ship Date"].dt.to_period("M"))["Sales"].sum()  # Time series analysis monthly
shipping_days
```

```
store["shipping_days"] = (store["Ship Date"] - store["Order Date"]).dt.days   # average shipping by category
shipping_days = store.groupby("Category")["shipping_days"].mean()
shipping_days
```

```
shipping_days.plot(kind = "pie" , autopct = "%1.1f%%")
plt.title("Average Shipping Time")
plt.show()
```
